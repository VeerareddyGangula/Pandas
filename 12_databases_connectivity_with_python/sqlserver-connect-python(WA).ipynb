```python
import pandas as pd
```

```python
from sqlalchemy import create_engine,text

# Replace server_name with the name of your SQL Server instance
server_name = ' mention server name'
database_name = 'mention database name'

# Create the connection string
connection_string = f'mssql+pyodbc://@{server_name}/{database_name}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server'

# Create the engine
engine = create_engine(connection_string)

con = engine.connect()
```

```python
res = con.execute(text("select * from geo where Geo ='India'"))
for row in res:
    print(row)
    
```
